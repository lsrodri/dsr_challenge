```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer
```

```python
# Load the dataset from the subfolder
df_features = pd.read_csv("../src/data/raw/dengue_features_train.csv")
df_target = pd.read_csv("../src/data/raw/dengue_labels_train.csv")

print(df_features.head())
```

```python
print(df_target.head())
```

```python
# Merge on "city", "year", and "weekofyear"
df = pd.merge(df_features, df_target[["city", "year", "weekofyear", "total_cases"]],
              on=["city", "year", "weekofyear"],
              how="left")
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# impute missing values with simple imputer

# Select only numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Create and apply the imputer
imputer = SimpleImputer(strategy='mean')
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
```

```python
df.isnull().sum()
```

```python
df.to_csv("full_data_frame.csv", index=False)
```
